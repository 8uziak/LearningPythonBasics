```python
import random

def ran_word():
    
    
    with open('hangman.txt', 'r') as f:
        
        f_random = random.randint(1,267751)
        
        for i in range(1,f_random): 
            
            f.readline()
            
        return f.readline()
```

```python
def guess_word(ran_word):
    
    word_to_guess = ran_word
    word_to_guess_len = len(word_to_guess)
    y_word = list((('_') * (word_to_guess_len - 1)))
    
    hangman_1 = "---"
    hangman_2 = '\n | ' + '\n | ' + '\n | ' + '\n | ' + '\n | ' + '\n | ' + "\n---" 
    hangman_3 = '\n  ________''\n |' + '\n | ' + '\n | ' + '\n | ' + '\n | ' + '\n | ' + "\n---" 
    hangman_4 = '\n  ________''\n |       |' + '\n |       O' + '\n | ' + '\n | ' + '\n | ' + '\n | ' + "\n---"
    hangman_5 = '\n  ________''\n |       |' + '\n |       O' + '\n |       |' + '\n | ' + '\n | ' + '\n | ' + "\n---"
    hangman_6 = '\n  ________''\n |       |' + '\n |       O' + '\n |      /|\ ' + '\n |      / \ ' + '\n | ' + '\n | ' + "\n---"
    
    hangman_lst = ('', hangman_1, hangman_2,hangman_3, hangman_4,hangman_5,hangman_6)
    

    game_on = True
    
    n = 0 
    wrong_ = 0
    
    print(word_to_guess)
    print(y_word)
    
    while game_on:
        
        
        if wrong_ == 6:
            print("**** You lost! ****")
            user_inp = input('Do you want to play again?(yes, no): ')
            if user_inp == 'no':
                game_on = False 
                word_to_guess = ran_word
            else: 
                wrong_ = 0
                n = 0
                word_to_guess = ran_word
                word_to_guess_len = len(word_to_guess)
                y_word = list((('_') * (word_to_guess_len - 1)))
                print(word_to_guess)
                print(y_word)
        else:    
            
            user_ = input("\nGuess the letter: ")
            
            for i in range(1,word_to_guess_len):
                if user_ == word_to_guess[n]:
                    y_word[n] = user_
                    n += 1
                else:    
                    n += 1
                    
                    
            if not user_ in y_word:
                wrong_ += 1
                print("\n** Wrong! **\n\n")
                print(hangman_lst[wrong_])

            n = 0

            print(y_word)
            print(wrong_)
            print("\n")

            if y_word == list(word_to_guess[:-1]):
                print("Congratulations! You won the game!")
                user_ = input("Do you want to play again?: ")
                if user_ == "no":
                    game_on = False
            
          
    
```
